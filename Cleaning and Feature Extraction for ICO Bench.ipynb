```python
import json
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
import numpy as np
from datetime import datetime
```

```python
f = open('ico_bench_data_cleaned.json', 'r')
ico_data = json.loads(f.read())
df = pd.DataFrame.from_dict(ico_data)
f.close()
df
```

```python
good_data = [e for e in ico_data if 'raised' in e and 'hard_cap' in e]
len(good_data)
```

```python
a = [item for item in good_data if len(item['raised']) == 0]
a
```

```python
mlb = MultiLabelBinarizer()
df = df.join(pd.DataFrame(mlb.fit_transform(df.pop('accepting')), columns=mlb.classes_, index=df.index))
df = df.join(pd.DataFrame(mlb.fit_transform(df.pop('categories')), columns=mlb.classes_, index=df.index))
```
